## Lists and Dictionaries

First we look at some built in Python data structures: lists and dictionaries. 

A list is a sequence of things, unlike strongly typed languages (Java, C#) a list can contain a mixture of different types - there is no type for a list of integers or a list of lists.   Here are some lists:

```python
ages = [12, 99, 51, 3, 55]
names = ['steve', 'jim', 'mary', 'carrie', 'zin']
stuff = [12, 'eighteen', 6, ['another', 'list']]
```

1. write code to print the first and third elements of each list

3. write a for loop that prints each element of the 'names' list

```python
lst = [ages,names,stuff]
def firstthird(l):
    for i, item in enumerate(l):
        if i == 0 or i == 2:
            yield item


def notfirst(l):
    for i,item in enumerate(l):
        if i !=0:
            yield item
            
for i in lst:
    print(eval('list(firstthird(i))'))            
for i in lst:
    print(eval('list(notfirst(i))'))
```

```python
#Add a new column, if column exists then old one will be overwritten
edu['ValueNorm'] = edu['Value']/edu['Value'].max()
edu.head()
```

```python
#Remove columns from df. removes rows if axis = 0, removes columns if axis = 1
#in pandas all functions that change the  contents of a DataFrame,
#such as the drop fn will return a copy of the modified data, instead of overwriting the df.
#If you do not want to keep the old values, you can set the keyword inplace to True.

edu.drop('ValueNorm',axis=1,inplace = True)
edu.head()
```

```python
#add new row, You must be aware to setting the ignore_index flag in the append method to True, 
#otherwise the index 0 is given to this new row, which will produce an error if it already exists:

edu = edu.append({'TIME':2000, 'Value':5.00,'GEO':'a'}, ignore_index = True)
edu.tail()
```

```python
#drops the last column in the row which is defined by max(edu.index)
edu.drop(max(edu.index),axis=0,inplace=True)
edu.tail()
```

```python
#drop() can also be used to remove missing values such as null values:
eduDrop = edu.dropna()
eduDrop.head()
```

```python
#to specify which rows, how = 'any' and to specify columns, subset = ['Value']

eduDrop = edu.dropna(how = 'any', subset = ['Value'])
eduDrop.head()
```

```python
#instead of removing rows containing Nan, we can fill them with another value using fillna():
eduFilled = edu.fillna(value = {'Value':0})
eduFilled
```

```python
#Sorting if we want to see top 5 in terms of Value, we can do this:

edu.sort_values(by='Value',ascending = False, inplace=True)
edu.head()
```

```python
#Grouping data, group based on criteria. To have a proper df as a result, we msut apply an aggregation function.
#For this example, if we want a df showing the mean of the values for each country over all the years,
#we can obtain it by grouping according to country and using the mean function as the aggregation method for each group.
# The result would be a DataFrame with countries as indexes and the mean values as the column:
#finds the average 'value' based on the GEO, average is measured by the values across all the time occurences.
group = edu[['GEO','Value']].groupby('GEO').mean()
group.head()
```

```python
#Rearranging Data: We can arrange our data using the pivot_table function.
# Here, we can specify which columns will be the new indexes, the new values and the new columns.
#Imagine that we want to transform our df to a spreadsheet-like structure with the country names as the index,
#while the columns will be the years starting from 2006 and the values will be the previous Value column.

filtered_data = edu[edu['TIME']>2005]
pivedu = pd.pivot_table(filtered_data, values = 'Value',
                       index = ['GEO'],
                       columns = ['TIME'])
pivedu
```

```python
#now we can use the new index to select specific rows by label using the loc operator:

pivedu.loc[['Spain','Portugal'],[2006,2011]]
```

```python
#Ranking data We would like to know how each country is ranked by year.
#First we must clean up previous pivoted table so it only has real countries with real data.
#To do this, we drop the Euro area entries and shorten the Germany name entry using the rename function,
#also drop all the rows containing any NaN using dropna
#Now we perform the ranking using the rank fn. In our case, we use the first method,
#in which ranks are assigned in the order they appear in the array, avoiding gaps between ranking.
#These numbers represent their current ranking or standing at the specific year which is measured by the 'value'
#ascending = False makes it so that the countries with the highest value will be awarded the rank 1.

pivedu = pivedu.drop(['Euro area (13 countries)',
                     'Euro area (15 countries)',
                     'European Union (25 countries)',
                     'European Union (27 countries)',
                     ],
                     axis=0)
pivedu = pivedu.rename(index={'Germany(until 1990 former territory of the FRG)':'Germany'})
pivedu = pivedu.dropna()
pivedu.rank(ascending = False, method = 'first').head()
```

```python
#If we want to make a global ranking taking into account all the years, we can sum up all the columns and rank the result.
#Then we can sort the resulting values to retrieve the top five countries for the last 6 years, in this way:
#the method keyword argument in the rank function specifies how items that compare equals receive ranking. 
#dense, items that compare equals receive the same ranking number.

totalSum = pivedu.sum(axis=1)
totalSum.rank(ascending = False, method = 'dense').sort_values()
```

```python
import matplotlib
import matplotlib.pyplot as plt
#Plotting: If we want to plot the accumulated values for each country over the last 6 years,
#we can take the Series obtained in the previous example and plot it directly by calling the plot fn:

totalSum = pivedu.sum(axis = 1).sort_values(ascending = False)
totalSum.plot(kind = 'bar', style = 'b', alpha = 0.8, title = 'Total Values for Country')
totalSum

#If we want highest to lowest, we need to use sort_values to sort the values in the series
#The parameter style refers to the style properties of the graphic, in our case, the colour b = blue
#The alpha channel can be modified adding a keyword
#parameter alpha with a percentage, producing a more translucent plot
```

```python
#It is also possible to plot a df directly. In this case each column is treated as a separated Series.
#For example, instead of printing the accumulated value over the years, we can plot the value for each year.

my_colours = ['b','r','g','y','m','c']
ax = pivedu.plot(kind = 'barh',
                stacked = True,
                color = my_colours)
ax.legend(loc = 'center left', bbox_to_anchor = (1,.5))

#kind = bar type, stacked = stacked bar, color = use a different colour for each year, 
#When several series are shown in a plot, a legend is created for identifying each one.
#The name of each series is the name of the column in the df. 
#by default, the legend goes inside the plot area. If we want to change this, we can use the legend function
#by using the loc keyword, we can set the relative position of the legend with respect to the plot.
#With bbox_to_anchor we can set an absolute position with respect to the plot, allowing us to put the legend outside the graph.
```
